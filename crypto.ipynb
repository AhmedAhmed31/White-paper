```python
from googlesearch import search 
import urllib.request
import pandas as pd
```

Cryptocurrency names list 

```python
file='crypto.csv'
def readnames(file):
    df=pd.read_csv(file)
    df=pd.DataFrame(data=df)
    crypto=df['name']
    return crypto
```

```python
readnames(file).head()

```

```python
def querysearch(crypto=readnames(file)):
    query=[]
    text="white parper pdf"
    for i in crypto:
        query.append( i.strip() +' '+ text )
    return query 
```

```python
query=querysearch()
query
```

```python
# to search
token=[]
for n in query:
    for j in search(n, tld="co.in", num=10, stop=1, pause=2): 
        token.append(j)
        print(j) 
```
